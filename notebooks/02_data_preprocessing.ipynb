```python
import re
import pandas as pd

pd.set_option('display.max_colwidth', 200)

df = pd.read_csv("../dataset/captions.txt");
```

```python
def clean_caption(text):

    text=text.lower()

    text=re.sub(
        r"[^a-z ]",
        "",
        text
    )

    text=text.strip()

    return text
```

```python
df["caption"]=(
df["caption"]
.apply(clean_caption)
)

print(df["caption"].iloc[0])
```

```python
df["caption"]=(
"<start> "
+
df["caption"]
+
" <end>"
)
```
